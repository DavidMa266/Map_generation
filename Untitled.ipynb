```python
import numpy as np
import pandas as pd
from sklearn.cluster import SpectralClustering
import csv
from scipy.sparse import csgraph
import matplotlib.cm as cm
import sklearn.decomposition as skd
from sklearn import manifold
from sklearn import cluster
from scipy.sparse.csgraph import laplacian
from numpy.linalg import eig
from sklearn.decomposition import PCA 
import numpy.linalg
from sklearn.cross_decomposition import CCA
import matplotlib.pyplot as plt
```

```python
def load_csv(destination, types):
    return np.array(pd.read_csv(destination,header=None)).astype(types)
```

```python
bad_result = load_csv('data_512_3.csv', float)
```

```python
i = 0
z = np.zeros((512,512))
for i in range(0, 512):
    for j in range(0,512):
        z[i][j] = bad_result[i * 512 + j][2]
bad_result = 0
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np

```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
```

```python
X, Y = np.meshgrid(np.arange(0,512,1), np.arange(0,512,1))
print(X)
print(Y)
```

```python
surf = ax.plot_surface(X, Y, np.asarray(z), rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```

```python
ax.set_zlim(-1.01, 1.01)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
